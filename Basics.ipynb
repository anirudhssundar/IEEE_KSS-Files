## What is an image?
>In its simplest terms, an image is a visual perception of a two-dimensional object or subject  
>A digital image however, is a numeric representation of the same object  
>A digital image is made up of a number of 'pixels'  
> Each pixel being nothing but a binary number  

In all our discussions henceforth, we deal only with digital images

```python
# Let's create our own image!!!

import numpy as np
import matplotlib.pyplot as plt

# Create a 16x16 matrix of 0s
image = np.zeros((16,16),dtype = 'uint8')
rows, cols = image.shape
for i in range(rows):
    for j in range(cols):
        image[i,j] = (i+j)%2 # Assign alternating 0s and 1s
# Print out our image
print(image)
```

Well, the image is kind of raw, isn't it?  
We don't really see numbers in real life  
What we really see is the pixels  
The numbers just represent the intensity of the pixels  

With the help of matplotlib we can plot the pixel values to make more sense of the above numbers

```python
%matplotlib inline
plt.imshow(image,cmap = 'gray')
plt.show()
```

There, now that's more like it!  
All the 0s are black squares and the 1s are white squares  
What you see above is a binary image  
That is, the pixels can only take binary values-- 0 or 1

## Color Spaces

The first image we viewed was in black and white  
We call such images **Binary Images**

The next image we viewed (Lena) had values ranging from 0 to 255  
Such images are called **Grayscale Images**

Now let's look at a third class of images that's much more appealing - **Color Images**

```python
img = cv2.imread('Images/lena.jpg',1)

plt.imshow(img)
plt.show()
```

But wait something's off isn't it?  
No human being is blue in color  
Well, this is because of a peculiarity in the way images are interpreted by OpenCV and Matplotlib  

To understand this we first need to look at the representation of color images

```python
print("dimensions = " + str(np.shape(img)))
```

As seen above, the image has 3 dimensions  
The first two dimensions represent the pixel intensity values  
Can you guess what the third channel is?

They are in fact the 3 primary colors - <span style="color:red">red</span>, <span style="color:green">green</span> and <span style="color:blue">blue</span>

```python
print("blue channel = ")
print(img[:,:,0])

print("green channel = ")
print(img[:,:,1])

print("red channel = ")
print(img[:,:,2])
```

### So what caused the weirdness in the color image we loaded?

OpenCV loads images in the order - Blue, Green, Red (BGR)  
Matlab and subsequently matplotlib interpret images as Red, Green, Blue  
So let's try switching these and see what happens!

```python
# A little snippet to swap the first and last channels
img2 = img[:,:,::-1]
```

```python
plt.imshow(img2)
plt.show()

print("red channel = ")
print(img2[:,:,0])

print("green channel = ")
print(img2[:,:,1])

print("blue channel = ")
print(img2[:,:,2])
```

Fortunately for us OpenCV comes with a built in command to do this "switching"  
It's nothing but cv2.cvtColor

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```
